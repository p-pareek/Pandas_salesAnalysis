# Exploratory data analysis

what is the overall sales trend?


```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#getting month-year from the datset
df['month_year']=df['order_date'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
#grouping month year 
df_monthly_trend=df.groupby('month_year').sum()['sales'].reset_index()
df_monthly_trend
```

```python
#setting the figure size
plt.figure(figsize=(15,6))
plt.plot(df_monthly_trend['month_year'],df_monthly_trend['sales'],color='red')
plt.xticks(rotation='vertical',size=8)
plt.show()
```

# which are the top 10 products by sales

```python
df.head()
```

```python
#grouping the product_name column
prod_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
prod_sales
```

```python
#sorting prod-sales column
prod_sales=prod_sales.sort_values('sales',ascending=False)
prod_sales
```

```python
#top 10 product by sales
top10_prodbysales=prod_sales[:10]
top10_prodbysales
```

# Which are the Most Selling Products?

```python
#grouping product name
most_sell_prod=pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
most_sell_prod=most_sell_prod.sort_values('quantity',ascending=False)
```

```python
most_sell_prod[:10]
```

# Most preferred ship mode 

```python
#setting figure ssize
plt.figure(figsize=(10,8.5))
#plotting shipmode
sns.countplot(df['ship_mode'])
```

# which are the most profitable category ans sub category 

```python
#grouping category and subcategory
cat_subcat_profit=pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
cat_subcat_profit=cat_subcat_profit.sort_values(['category','profit'],ascending=False)
cat_subcat_profit
```
